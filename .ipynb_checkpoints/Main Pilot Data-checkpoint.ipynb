```python
anxiety.describe()
# PRCSPRE: Baseline score ontained before their VR experience; Higher score --> less confidence, more fear of public speaking.
    # Scale min 30 and max 180
# STAI: Score obtained before and after each experience, experimental and follow-up; Higher score --> higher state anxiety.
    # Scale min 0 and max 30
# IAT: Positive d-scores: support a stronger association between 'Me-Liked' and 'Others-Rejected'
    # Negative d-scores: support a stronger association between 'Others-Liked' and 'Me-rejected'
    # Negative Score--> more fear of negative evaluation/rejection
# BFNE: Baseline score obtained before their VR experience; Higher score --> higher Fear of negative evaluation
    # Scale min 16 and max 60 (Avg population 28.81 (4.78))
# SPIC: Index of competence regarding future planned talk (follow-up session)
    # The questionnaire is completed right after experimental session 1 (a week before session 2)
    # Higher scores --> less perceived competence to complete the task; Min avg 15 max 105
    # Between-groups analysis
# Anxiety: Anxiety related to the follow-up public speech (session 2) during the preparation period (1 week between sessions)
    # Higher score --> Higher anxiety; Min 0 Max 27
```
